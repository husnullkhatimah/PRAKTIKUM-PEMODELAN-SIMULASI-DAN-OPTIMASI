## **Import Library**

```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.linear_model import LinearRegression

from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
```

## **Membaca Dataset**
Dataset yang digunakan misalnya dataset harga rumah.

```python
data = pd.read_csv('https://raw.githubusercontent.com/GeostatsGuy/GeoDataSets/master/1D_Porosity.csv')
data.head()
```

## **Informasi Dataset**

```python
data.info()
```

## **Statistik Deskriptif**

```python
data.describe()
```

## **Visualisasi Hubungan Variabel**

```python
sns.scatterplot(x='Depth', y='Nporosity', data=data)
plt.show()
```

## **Regresi Linier Sederhana**

Model regresi:

Y = β0 + β1X + ε

## **Menentukan Variabel**

```python
X = data[['Depth']]
y = data['Nporosity']
```

## **Membuat Model Regresi**

```python
model = LinearRegression()
model.fit(X, y)
```

## **Melihat Koefisien Model**

```python
print('Intercept:', model.intercept_)
print('Koefisien:', model.coef_)
```

## **Membuat Visualisasi Model**

```python
y_pred = model.predict(X)
print(y_pred)
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Depth', y='Nporosity', data=data, label='Data Asli')
plt.plot(X, y_pred, color='purple', label='Garis Regresi Orde 1')
plt.title('Visualisasi Model Regresi Linier Sederhana')
plt.xlabel('Depth')
plt.ylabel('Nporosity')
plt.legend()
plt.grid(True)
plt.show()
```

## **Scatter Plot Regresi Linear (Regresi Sederhana)**

Penjelasan:

Visualisasi ini menunjukkan model regresi linier sederhana antara Depth (sumbu X) dan Npnosti (sumbu Y). Titik-titik biru adalah data asli, sementara garis merah adalah garis regresi yang merepresentasikan hubungan linier antar kedua variabel.

Dari grafik terlihat bahwa garis merah menurun dari kiri ke kanan, yang mengindikasikan hubungan negatif antara Depth dan Npnosti. Artinya, semakin besar nilai Depth (kedalaman), maka nilai Npnosti cenderung semakin kecil. Sebaran titik-titik data yang relatif dekat dengan garis merah menunjukkan bahwa model regresi ini cukup akurat dalam menjelaskan hubungan tersebut, sehingga Depth dapat diandalkan sebagai prediktor untuk Npnosti.memprediksi data.

### **Prediksi**

```python
y_pred = model.predict(X)
print(y_pred)
```

## **Mean Absolute Error (MAE) , MEan Squared Error (MSE) dan Root Mean Square Error (RMSE)**

```python
mae = mean_absolute_error(y, y_pred)
mse = mean_squared_error(y, y_pred)
rmse = np.sqrt(mean_squared_error(y, y_pred))


print('MAE:', mae)
print('MSE:',mse)
print('RMSE:',rmse)
```

## **Regresi Polynomial**

Regresi polynomial merupakan pengembangan dari regresi linier yang digunakan untuk memodelkan hubungan nonlinier antara variabel independen dan variabel dependen. Jika hubungan antara variabel tidak dapat dijelaskan dengan garis lurus, maka regresi polynomial dapat digunakan dengan menambahkan pangkat dari variabel independen ke dalam model.

Screenshot 2026-03-07 093826.png

## **Membuat model orde 2**

```python
from sklearn.preprocessing import PolynomialFeatures
poly = PolynomialFeatures(degree=2)
X_poly = poly.fit_transform(X)
X_poly
```

```python
#Membuat model orde 2
model2 = LinearRegression()
model2.fit(X_poly, y)
```

```python
y_pred2 = model2.predict(X_poly)
print(y_pred2)
```

```python
print('Intercept:', model2.intercept_)
print('Koefisien:', model2.coef_)
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Depth', y='Nporosity', data=data, label='Data Asli')
plt.plot(X, y_pred, color='purple', label='Garis Regresi Orde 1')
plt.plot(X, y_pred2, color='pink', label='Garis Regresi Orde 2')
plt.title('Scatter Plot Regresi Linear Sederhana (Depth vs Nporosity)')
plt.xlabel('Depth')
plt.ylabel('Nporosity')
plt.legend()
plt.grid(True)
plt.show()
```

## **Scatter Plot Regresi Polynomial Orde 2**
Penjelasan:

Grafik scatter plot ini menggambarkan hubungan antara variabel kedalaman (Depth) dan porositas (Nporosity) dengan membandingkan dua model pendekatan regresi. Secara keseluruhan, terlihat adanya tren positif yang lemah di mana nilai porositas cenderung meningkat seiring bertambahnya kedalaman, meskipun sebaran data aslinya sangat bervariatif dan tidak membentuk garis lurus yang sempurna. Penggunaan Garis Regresi Orde 1 (ungu) menunjukkan model linear sederhana yang mengasumsikan kenaikan konstan secara konsisten dari awal hingga akhir data. Namun, model ini tampak terlalu kaku karena tidak mampu menangkap fluktuasi data pada rentang kedalaman tertentu.

Di sisi lain, Garis Regresi Orde 2 (pink) memberikan pendekatan polinomial yang lebih fleksibel dengan bentuk kurva melengkung. Model ini mencoba menyesuaikan diri dengan pola data yang terlihat meningkat tajam di awal, lalu mulai melandai atau mengalami sedikit penurunan pada kedalaman menengah (sekitar nilai 4 hingga 8). Meskipun model orde 2 terlihat sedikit lebih mendekati pola persebaran titik-titik data dibandingkan model linear, keberadaan data pencilan (outliers) yang cukup signifikan—terutama penurunan drastis pada kedalaman 6 hingga 8—menunjukkan bahwa faktor kedalaman saja mungkin tidak cukup kuat untuk menjelaskan seluruh perubahan nilai porositas dalam sampel ini.

```python
mae = mean_absolute_error(y, y_pred2)
mse = mean_squared_error(y, y_pred2)
rmse = np.sqrt(mean_squared_error(y, y_pred2))


print('MAE:', mae)
print('MSE:',mse)
print('RMSE:',rmse)
```
